 2: What are the key assumptions of Simple Linear Regression?

 Ans:- 🔑 1. Linearity
Assumption: There is a linear relationship between the independent variable (X) and the dependent variable (Y).

Why it matters: If the relationship is not linear, predictions will be inaccurate.

📈 Check with: Scatter plot of X vs Y or residual plot.

🔑 2. Independence of Errors
Assumption: The residuals (errors) are independent of each other.

Why it matters: If errors are correlated (e.g., in time series), estimates may be misleading.

📉 Check with: Durbin-Watson test (especially for time series data).

🔑 3. Homoscedasticity (Constant Variance of Errors)
Assumption: The variance of the residuals is constant across all levels of X.

Why it matters: If variance changes (heteroscedasticity), the model may underestimate or overestimate the uncertainty.

📊 Check with: Plot of residuals vs fitted values.

🔑 4. Normality of Errors
Assumption: The residuals are normally distributed.

Why it matters: Important for making valid confidence intervals and hypothesis tests.

📋 Check with: Histogram or Q-Q plot of residuals.

🔑 5. No Perfect Multicollinearity (Not applicable to simple linear regression)
This applies to multiple linear regression, not simple linear regression, since there’s only one independent variable.


3: What is heteroscedasticity, and why is it important to address in regression
models?


Ans:- 🔍 Definition:
Heteroscedasticity occurs when the variance of the residuals (errors) in a regression model is not constant across all levels of the independent variable(s).

In contrast, homoscedasticity (a key assumption of regression) means the residuals have constant variance.

📈 Visual Example:
In a scatter plot of residuals vs. predicted values:

Homoscedasticity → Residuals are spread evenly.

Heteroscedasticity → Residuals form patterns like cones or funnels (widening or narrowing spread).

⚠️ Why Is Heteroscedasticity a Problem?
Invalid Standard Errors
→ It causes incorrect estimation of standard errors, leading to unreliable:

t-tests

p-values

confidence intervals

Unreliable Hypothesis Tests
→ You might incorrectly conclude that a variable is significant (Type I or II errors).

6: Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]

Ans:-



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape for sklearn
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and train the model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Print the model parameters
print("Intercept (β₀):", model.intercept_)
print("Slope (β₁):", model.coef_[0])

# Plotting
plt.scatter(X, Y, color='blue', label='Actual data')
plt.plot(X, Y_pred, color='red', label='Regression line')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```

 7: Fit a Multiple Linear Regression model on this sample data:
● Area = [1200, 1500, 1800, 2000]
● Rooms = [2, 3, 3, 4]
● Price = [250000, 300000, 320000, 370000]

Ans:-

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Input features (Area, Rooms)
X = np.array([
    [1200, 2],
    [1500, 3],
    [1800, 3],
    [2000, 4]
])

# Target variable (Price)
Y = np.array([250000, 300000, 320000, 370000])

# Create and train the model
model = LinearRegression()
model.fit(X, Y)

# Print model coefficients
print("Intercept (β₀):", model.intercept_)
print("Coefficients (β₁ for Area, β₂ for Rooms):", model.coef_)

# Predict prices using the model
Y_pred = model.predict(X)
print("\nPredicted Prices:", Y_pred)
```

8: Implement polynomial regression on the following data:
● X = [1, 2, 3, 4, 5]
● Y = [2.2, 4.8, 7.5, 11.2, 14.7]

Ans:-

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict using the model
Y_pred = model.predict(X_poly)

# Print model coefficients
print("Intercept (β₀):", model.intercept_)
print("Coefficients:", model.coef_)

# Plot
plt.scatter(X, Y, color='blue', label='Actual data')
plt.plot(X, Y_pred, color='red', label='Polynomial Regression Line')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Regression (Degree 2)")
plt.legend()
plt.grid(True)
plt.show()


```

9: Create a residuals plot for a regression model trained on this data:
● X = [10, 20, 30, 40, 50]
● Y = [15, 35, 40, 50, 65]


Ans:-

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Train simple linear regression model
model = LinearRegression()
model.fit(X, Y)

# Predictions and residuals
Y_pred = model.predict(X)
residuals = Y - Y_pred

# Plot residuals
plt.scatter(X, residuals, color='purple')
plt.axhline(y=0, color='black', linestyle='--')  # Zero residual line
plt.xlabel("X")
plt.ylabel("Residuals (Y - Ŷ)")
plt.title("Residuals Plot")
plt.grid(True)
plt.show()

# Optional: Print residuals
print("Residuals:", residuals)
```

b. Use Weighted Least Squares (WLS)
Assign less weight to observations with larger residual variance.

This gives more reliable coefficient estimates.

c. Use Robust Standard Errors
Use heteroscedasticity-consistent standard errors (e.g., via statsmodels in Python with HC0, HC1).

These correct inference problems without changing the model itself.

🔍 2. Addressing Multicollinearity
Multicollinearity occurs when two or more features are highly correlated, inflating standard errors and making coefficients unstable.

✅ Steps:
a. Check Correlation Matrix or VIF
Calculate Variance Inflation Factor (VIF):

VIF > 5 or 10 = strong multicollinearity.

Also check feature correlation heatmaps.

b. Drop or Combine Features
Remove one of the highly correlated variables (e.g., drop either "area" or "lot size" if they are strongly correlated).

Or combine them using techniques like Principal Component Analysis (PCA).

c. Use Regularization Techniques
Use Ridge Regression (L2) or Lasso Regression (L1):

These penalize large coefficients and help reduce the impact of multicollinearity.

Lasso can also perform feature selection by shrinking some coefficients to 0.

📈 3. Model Evaluation and Validation
After making corrections:

Refit the model with corrected inputs.

Check residual plots again for heteroscedasticity.

Re-calculate VIF scores.

Use cross-validation to assess generalization.

Compare Adjusted R², RMSE, and MAE before and after.

🧠 Example:
Say you had the following features:

area, num_rooms, bathrooms, location_encoded

If area and num_rooms are highly correlated:

Drop one or combine into a new variable like area_per_room.

If variance of residuals increases with price:

Use log(price) as your target variable and interpret results accordingly.
